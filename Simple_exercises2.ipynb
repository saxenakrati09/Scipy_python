Diffusion using Random walk algorithm

Letusconsiderasimple1Drandomwalkprocess:ateachtimestepawalkerjumpsrightorleftwith equal probability.
We are interested in finding the typical distance from the origin of a random walker after t left or right jumps? We are going to simulate many â€œwalkersâ€ to find this law, and we are going to do so using array com- puting tricks: we are going to create a 2D array with the â€œstoriesâ€ (each walker has a story) in one direction, and the time in the other:

```python
n_stories = 1000 # number of walkers
t_max = 200 # time which we follow the walker

# We randomly choose all the steps 1 or -1 of the walk
t = np.arange(t_max)
#steps = 2 * np.random.randint(0, 1+1, (n_stories, t_max))
steps = np.random.choice([1,-1],(n_stories, t_max))
np.unique(steps)
```

```python
# We build the walk by summing steps along the time

positions = np.cumsum(steps, axis=1)
sq_distance = positions**2

# we get the mean of the axis of the stories
mean_sq_distance = np.mean(sq_distance, axis = 0)

```

```python
#plot the figure

plt.figure(figsize=(4,3))
plt.plot(t, np.sqrt(mean_sq_distance), 'g.', t, np.sqrt(t),'y-')
plt.xlabel(r"$t$")
plt.ylabel(r"$\sqrt{\langle (\delta x)^2 \rangle}$")
plt.tight_layout() # provide sufficient space for labels
```

Example using polynomials in Chebyshev basis for polynomials in range [-1, 1]

```python
x = np.linspace(-1, 1, 2000)
y = np.cos(x) + 0.3*np.random.rand(2000)
p = np.polynomial.Chebyshev.fit(x,y,90)

t = np.linspace(-1, 1, 200)
plt.plot(x,y,'r.')
plt.plot(t,p(t), 'g-', lw=3)
```

Write a script to compute Mandelbrot fractal

```python
def compute_mendelbret(N_max, some_threshold, nx, ny):
    # A grid of c-values
    x = np.linspace(-2, 1, nx)
    y = np.linspace(-1.5, 1.5, ny)
    
    c = x[:, np.newaxis]+1j*y[np.newaxis,:]
    # Mendelbret Iteration
    
    z = c
    for j in range(N_max):
        z = z**2 + c
    
    mendelbret_set = (abs(z) < some_threshold)
    
    return mendelbret_set

mendelbret_set = compute_mendelbret(50, 50., 601, 401)
plt.imshow(mendelbret_set.T, extent=[-2, 1, -1.5, 1.5])
plt.gray()
```
